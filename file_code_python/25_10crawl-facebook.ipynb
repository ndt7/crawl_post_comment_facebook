```python
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
import os

# ID_GROUP ="groups/1534645983449236"
# FANPAGE_LINK = "groups/338867744934119"  # hoi chi em sinh tai bv Phuong dong 
FANPAGE_LINK = "/groups/338867744934119"   # viem nam kho noi 
# ID_GROUP = "groups/561666328718461"   # nam khoa 
# ID_GROUP = "groups/338867744934119"   # phu khoa ha noi 
# ID_GROUP = "892534134455805"
ID_GROUP = "892534134455805"
# ID_GROUP = "groups/892534134455805"
# ID_GROUP = "groups/892534134455805"



# FANPAGE_LINK = "groups/" + ID_GROUP
PAGE_NUMBER = 10
COOKIE_PATH =           "C:\\Users\\ThinhNguen\\Desktop\\FaceBook\\crawl_post_comment_facebook\\file_data_cookies_facebook\\facebook.com_cookies.txt"
path_excel_post =       "C:\\Users\\ThinhNguen\\Desktop\\FaceBook\\crawl_post_comment_facebook\\post.xlsx"
path_excel_comment =    "C:\\Users\\ThinhNguen\\Desktop\\FaceBook\\crawl_post_comment_facebook\\comment.xlsx"
os.remove(path_excel_comment)
os.remove(path_excel_post)
post_list = []
for post in get_posts(FANPAGE_LINK, 
                    options={"comments": True, "reactions": None, "allow_extra_requests": True}, 
                    extra_info=None, pages=PAGE_NUMBER, cookies=COOKIE_PATH):
    post_list.append(post)


post_df_full = pd.DataFrame(columns = [])
for post in post_list:
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose()
    post_df_full = post_df_full.append(fb_post_df)
arr_npy = np.array(post_list) 

def save_comment(post_list, path_excel):
#   post_list = np.load(path_npy,allow_pickle=True) # load file npy =>nd_array
  post_list=post_list.tolist() # chuyeern ddinhj dangj sang lisy
  list_post=[]  #list post
  list_comment=[] #list comment
  for post in post_list: # lấy từng post 
      list_post.append(post['post_text']) # append từng post
      temp=[] # list các cooment của post đó
      if len(post['comments_full']) > 0: # nếu có comment
        for comment in post['comments_full']:
          
          s=""
          s+=comment['commenter_name']
          s+="****"
          for i in comment['comment_text']:
            if i.isalnum() or i.isdigit() or i.isnumeric() or i.isspace() :
              s+=i
          temp.append(s)
      list_comment.append(temp)
  index=[]
  detail_comment=[]

  for i in range (len( list_post)):
    for j in range(len(list_comment[i])):
      index.append(i)
      detail_comment.append(list_comment[i][j])
  df1 = pd.DataFrame(list(zip(index, detail_comment)),
                columns =['id', 'listcomment'])
  df1.to_excel(path_excel)

# =======================================

def save_post(post_list, path_excel):
  post_list=post_list.tolist() # chuyeern ddinhj dangj sang lisy
  
  list_post=[]  #list post
  list_comment=[] #list comment

  for post in post_list: # lấy từng post 
      s=""
      s+=post['username']
      s+="****"
      for i in post['post_text']:
        if i.isalnum() or i.isdigit() or i.isnumeric() or i.isspace() :
          s+=i
      list_post.append(s)

      temp=[] # list các cooment của post đó
      if len(post['comments_full']) > 0: # nếu có comment
        for comment in post['comments_full']:
          temp.append(comment['comment_text'])

      list_comment.append(temp)

  index=[] # id post
  for i in range (0,len( list_post)):# 
    index.append(i)
  df = pd.DataFrame(list(zip(index, list_post)),
                columns =['id', 'listpost'])
  df.to_excel(path_excel) # lưu df thành excel 

save_comment( arr_npy,path_excel_comment)
save_post( arr_npy,path_excel_post)

```
