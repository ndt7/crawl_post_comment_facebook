```python
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
```

# bắt đầu cào post từ 1 page của Facebook

```python
FANPAGE_LINK ="usuktimes"
FOLDER_PATH = "C:\\Users\\Admin\\Desktop\\crawlFacebook\\"
COOKIE_PATH = "C:\\Users\\Admin\\Desktop\\crawlFacebook\\facebook.com_cookies.txt"
path_excel_post = "C:\\Users\\Admin\\Desktop\\crawlFacebook\\222usuktimes.xlsx"
path_excel_comment = "C:\\Users\\Admin\\Desktop\\crawlFacebook\\222comment.xlsx"
PAGES_NUMBER = 10 
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK, 
                    options={"comments": True, "reactions": True, "allow_extra_requests": True}, 
                    extra_info=True, pages=PAGES_NUMBER, cookies=COOKIE_PATH):
    post_list.append(post)
```

```python
len(post_list)
arr_post_list = np.array(post_list)
```

```python
def save_post(post_list, path_excel):
  # post_list = np.load(path_npy,allow_pickle=True) # load file npy =>nd_array
  post_list=post_list.tolist() # chuyeern ddinhj dangj sang lisy
  
  list_post=[]  #list post
  list_comment=[] #list comment

  for post in post_list: # lấy từng post 
      s=""
      for i in post['post_text']:
        if i.isalnum() or i.isdigit() or i.isnumeric() or i.isspace() :
          s+=i
      list_post.append(s)

      # list_post.append(post['post_text']) # append từng post
      temp=[] # list các cooment của post đó
      if len(post['comments_full']) > 0: # nếu có comment
        for comment in post['comments_full']:
          temp.append(comment['comment_text'])
        # cmt = '\n'.join([comment['comment_text'] for comment in post['comments_full'] ])
      list_comment.append(temp)

  index=[] # id post
  for i in range (0,len( list_post)):# 
    index.append(i)
  df = pd.DataFrame(list(zip(index, list_post)),
                columns =['id', 'listpost'])
  df.to_excel(path_excel) # lưu df thành excel 
save_post(arr_post_list,path_excel_post)
```

```python
def save_comment(post_list, path_excel):
  # post_list = np.load(path_npy,allow_pickle=True) # load file npy =>nd_array
  post_list=post_list.tolist() # chuyeern ddinhj dangj sang lisy
  
  list_post=[]  #list post
  list_comment=[] #list comment

  for post in post_list: # lấy từng post 
      list_post.append(post['post_text']) # append từng post
      temp=[] # list các cooment của post đó
      if len(post['comments_full']) > 0: # nếu có comment
        for comment in post['comments_full']:
          
          s=""
          for i in comment['comment_text']:
            if i.isalnum() or i.isdigit() or i.isnumeric() or i.isspace() :
              s+=i
          temp.append(s)
          # temp.append(comment['comment_text'])
        # cmt = '\n'.join([comment['comment_text'] for comment in post['comments_full'] ])
      list_comment.append(temp)
  index=[]
  detail_comment=[]

  for i in range (len( list_post)):
    for j in range(len(list_comment[i])):
      index.append(i)
      detail_comment.append(list_comment[i][j])
  df1 = pd.DataFrame(list(zip(index, detail_comment)),
                columns =['id', 'listcomment'])
  df1.to_excel(path_excel)
save_comment(arr_post_list,path_excel_comment)
```
